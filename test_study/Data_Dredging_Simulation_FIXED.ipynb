# Data Dredging Simulation

This notebook demonstrates how running many hypothesis tests increases the number of false positives.

We simulate random data where **no real relationship exists**, but some tests still appear significant by chance.


```
# Step 1 — Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
```

## Step 2 — Function to simulate false positives

```
def simulate_false_positives(n_tests, alpha=0.05):
    false_count = 0
    for _ in range(n_tests):
        # generate independent random variables
        x = np.random.randn(100)
        y = np.random.randn(100)
        
        # compute correlation p-value
        _, p = stats.pearsonr(x, y)
        
        if p < alpha:
            false_count += 1
    return false_count
```

## Step 3 — Run simulation for different numbers of tests

```
tests_list = [10, 50, 100, 500]
results = []

for n in tests_list:
    fp = simulate_false_positives(n)
    results.append(fp)
    print(f"Tests: {n}, False positives: {fp}")
```

## Step 4 — Visualize the effect

```
plt.figure(figsize=(6,4))
plt.plot(tests_list, results, marker='o')
plt.xlabel("Number of Tests")
plt.ylabel("False Positives Found")
plt.title("Simulation of Data Dredging Effect")
plt.grid(True)
plt.show()
```
